```python
status_counts = data.groupby(['Marital Status', 'Status']).size().unstack(fill_value=0)

plt.figure(figsize=(10, 6))
status_counts.plot(kind='pie', subplots=True, autopct='%1.1f%%', startangle=90, cmap='Set3', legend=False)
plt.title('Comparison of Marital Status by Status')
plt.ylabel(None)

plt.text(-3.3, -1.5, 'Alive', color='blue', fontsize=12)
plt.text(-0.3, -1.5, 'Dead', color='red', fontsize=12)

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Menghitung persentase untuk setiap kategori
total_per_marital_status = data.groupby('Marital Status')['Grade'].count()
percentage_per_marital_status = data.groupby(['Marital Status', 'Grade']).size() / total_per_marital_status * 100
percentage_per_marital_status = percentage_per_marital_status.reset_index(name='Percentage')

plt.figure(figsize=(8, 6))

sns.barplot(x='Marital Status', y='Percentage', hue='Grade', data=percentage_per_marital_status)
plt.title('Marital Status vs Cancer Grade (Percentage)')
plt.xlabel('Marital Status')
plt.ylabel('Percentage')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Menghitung persentase untuk setiap kategori
total_per_race = data.groupby('Race')['Grade'].count()
percentage_per_race = data.groupby(['Race', 'Grade']).size() / total_per_race * 100
percentage_per_race = percentage_per_race.reset_index(name='Percentage')

plt.figure(figsize=(8, 6))

sns.barplot(x='Race', y='Percentage', hue='Grade', data=percentage_per_race)
plt.title('Race vs Cancer Grade (Percentage)')
plt.xlabel('Race')
plt.ylabel('Percentage')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Plot histogram for age
sns.histplot(data=data, x='Age', bins=20, kde=True, ax=axes[0])
axes[0].set_title('Distribution of Age')

# Plot histogram for tumor size
sns.histplot(data=data, x='Tumor Size', bins=20, kde=True, ax=axes[1])
axes[1].set_title('Distribution of Tumor Size')

# Show plot
plt.show()
```

## Label Encoder

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
encoder = LabelEncoder()

# Encode each feature
data['Race'] = encoder.fit_transform(data['Race'])
data['Marital Status'] = encoder.fit_transform(data['Marital Status'])
data['T Stage '] = encoder.fit_transform(data['T Stage '])
data['N Stage'] = encoder.fit_transform(data['N Stage'])
data['6th Stage'] = encoder.fit_transform(data['6th Stage'])
data['differentiate'] = encoder.fit_transform(data['differentiate'])
data['Grade'] = encoder.fit_transform(data['Grade'])
data['A Stage'] = encoder.fit_transform(data['A Stage'])
data['Estrogen Status'] = encoder.fit_transform(data['Estrogen Status'])
data['Progesterone Status'] = encoder.fit_transform(data['Progesterone Status'])
data['Status'] = encoder.fit_transform(data['Status'])

print(data)
```

## visualisasi untuk melihat seberapa pengaruh variabel T Stage 	N Stage	6th Stage terhadap Grade dan Tumor Size

```python
plt.figure(figsize=(16, 6))
plt.subplot(1, 2, 1)
sns.boxplot(x='T Stage ', y='Grade', hue='6th Stage', data=data, palette='coolwarm')
plt.title('Distribution of Grade by T Stage and 6th Stage')
plt.xlabel('T Stage')
plt.ylabel('Grade')
plt.xticks(rotation=45)

# Membuat subplot untuk memvisualisasikan pengaruh variabel T Stage, N Stage, dan 6th Stage terhadap Tumor Size
plt.subplot(1, 2, 2)
sns.boxplot(x='T Stage ', y='Tumor Size', hue='N Stage', data=data, palette='Set2')
plt.title('Tumor Size Distribution by T Stage and N Stage')
plt.xlabel('T Stage')
plt.ylabel('Tumor Size')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))

# Konversi kolom Estrogen Status dan Progesterone Status menjadi string
data['Estrogen Status'] = data['Estrogen Status'].astype(str)
data['Progesterone Status'] = data['Progesterone Status'].astype(str)

# Menggabungkan variabel Estrogen Status dan Progesterone Status dalam satu kolom untuk membuat plot yang lebih rapi
data['Hormonal Status'] = data['Estrogen Status'] + '_' + data['Progesterone Status']

# Membuat barplot untuk melihat pengaruh variabel Estrogen Status dan Progesterone Status terhadap Grade
sns.barplot(x='Hormonal Status', y='Grade', data=data, palette='coolwarm', ci=None)
plt.title('Effect of Hormonal Status on Grade')
plt.xlabel('Hormonal Status')
plt.ylabel('Grade')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
